Country and Capital Data

Now that I have the Pokemon name data I will use Pandas to scrap the names of all the Countries and Capital cities in the world. Once I have this I will convert the data into lists so that they can be used in the game. 

```python
import numpy as np
import pandas as pd

url = "https://www.boldtuesday.com/pages/alphabetical-list-of-all-countries-and-capitals-shown-on-list-of-countries-poster"
dframe_list = pd.io.html.read_html(url)
```

```python
dframe = dframe_list[0]
newheader =dframe.iloc[0]
dframe = dframe[1:]
dframe.columns = newheader
dframe.head()
```

```python
dframe.isna().sum()
```

Cleaning Data

At this point I can see that the data contains the letters of the alphabet to break up the Countries. After counting the null values in the Capital column and seeing that the number is smaller than the total number of letters in the alphabet I am happy to remove the null values, which will then remove the singluar letters

```python
dframe.dropna(subset = ["CAPITAL"],inplace =True)
dframe.head()
```

It worked!

```python
capitals_list = dframe['CAPITAL'].tolist()
country_list = dframe['COUNTRY'].tolist()
```

Now that I have separate lists for the Pokemon, Capital Cities and Countries data I can begin programming the game. TO start I will separate key sections in the game to make the code more readable as I go along

```python
##This block takes in the users choice of game and generates a random word in the respective section. 
##It also removes space characters in the word and replaces them with a slash character

def word_selector(list_name,num):
    global answer_list
    global guess
    global answer
    answer = list_name[num].lower()
    answer_list = list(answer)
    guess = ["_"]*len(answer_list)
    if " " in answer_list:
        for i in range(len(guess)):
            if answer_list[i] == " ":
                guess[i] = "/"
```

```python
## This block checks that there are still characters that need to be guessed
def completed_checker(guess):
    if "_" in guess:
        return True
    else:
        return False
```

```python
## This block prints game data following every turn
def print_info(guess, guessed_letters, remaining_lives, goes_taken):
    print("Hidden word - " , guess)
    print("Used letters - ",guessed_letters)
    print("Remaining lives - ", remaining_lives)
    print("Goes taken - ", goes_taken)
    print("Take another go!\n")
    
```
